```python
url = 'https://raw.githubusercontent.com/leekyuyoung20230313/pythone/main/datas/stock-data.csv'
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(url)
df.head()
```

```python
# 시계열 객체로 변경  to_datetime()
df['new_date'] = pd.to_datetime(df['Date'])
```

```python
df = df.set_index('new_date') 
```

```python
# timestamp -> Priod 로 변환  to_period() 함수  일정 기간을 나타내는 Period 객체
datas = ['2020-01-01','2020-05-05','2020-05-01','2023-04-20']
ts_datas = pd.to_datetime(datas)
ts_datas
```

```python
ts_datas.to_period(freq='D')
```

```python
ts_datas.to_period(freq='M')
```

```python
# A는 1년의 기간을 나타낸다 1년의 끝은 12월
ts_datas.to_period(freq='A') 
```

```python
# 시계열 데이터 만들기
# timestamp 배열
ts_ms = pd.date_range(start = '2023-01-01',
              end=None,
              periods=6,    #생성할 timestamp의 개수
              freq='MS',    # 시간간격(MS: 월의 시작일)
              tz = 'Asia/Seoul' # 시간대
              )
ts_ms
```

```python
ts_ms = pd.date_range(start = '2023-01-01',
              end=None,
              periods=6,    #생성할 timestamp의 개수
              freq='3M',    # 시간간격(MS: 월의 시작일)
              tz = 'Asia/Seoul' # 시간대
              )
ts_ms
```

```python
ts_ms = pd.date_range(start = '2023-01-01',
              end=None,
              periods=6,    #생성할 timestamp의 개수
              freq='M',    # 시간간격(MS: 월의 시작일)
              tz = 'Asia/Seoul' # 시간대
              )
ts_ms
```
