```python
import matplotlib.pyplot as plt
```

```python
X = range(100)
Y = range(100)
plt.plot(X,Y)
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(5,5))
X = range(100)
Y = range(100)
plt.plot(X,Y)
plt.title('A product')
plt.xlabel('weight')
plt.ylabel('height')
plt.show()
```

```python
# 화면 분할 2가지 방식
plt.figure(figsize=(10,3))

plt.subplot(1,2,1)
X=range(100)
Y=range(100)
plt.title('A')
plt.plot(X,Y)

plt.subplot(1,2,2)
X=range(100)
Y=range(100)
plt.title('B')
plt.plot(X,Y)

plt.show()

```

```python
import numpy as np
X=range(100)
Y=np.array(range(100))/2
plt.plot(X,Y)

Y2 = range(100)
plt.plot(X,Y2)

plt.legend(['A','B'])  # 범주

plt.show()
```

```python
X = range(100)
Y1 = [ np.cos(i) for i in X ]
Y2 = [ np.sin(i) for i in X ]

plt.plot(X,Y1)
plt.plot(X,Y2)
plt.show()
```

```python
# 화면분할 2  - 배열 및 벡터의 인덱스와 동일한 방법으로 사용
fig, ax = plt.subplots(2,2)

ax[0,0].plot(X,Y1)
ax[0,1].plot(X,Y2)
ax[1,0].plot(X,Y1)
ax[1,1].plot(X,Y2)
plt.show()
```

```python
# 색상
X = range(100)
Y1 = range(100)
Y2 = np.array(X)+100
# RGB 형식 및 팔레트 형식
plt.plot(X,Y1,color='#8569B2')
plt.plot(X,Y2,c='black')
plt.show()
```

```python
# 색상
X = range(100)
Y1 = range(100)
Y2 = np.array(X)+100
# RGB 형식 및 팔레트 형식
plt.plot(X,Y1,color='#8569B2', linestyle='dashed')
plt.plot(X,Y2,c='black', ls = 'dotted')
plt.show()
```

```python
# 범부
X = range(100)
Y1 = range(100)
Y2 = np.array(X)+100
# RGB 형식 및 팔레트 형식
plt.plot(X,Y1,color='#8569B2', linestyle='dashed',label='dashed plot')
plt.plot(X,Y2,c='black', ls = 'dotted', label = 'dotted plot')
plt.legend(shadow = True)
plt.show()
```

```python
import random
X_R = list(range(100))
random.shuffle(X_R)
print(X_R)
plt.plot(X_R,Y1)
plt.show()
```

```python
from pandas.core.arrays.sparse import dtype
# 시계열 데이터
# 지난시간 숙박업 년도별 영업과 폐업현황을 plt.plot()함수를이용해서 그려보세요.
import pandas as pd
import numpy as np
fpath = '/content/drive/MyDrive/Colab Notebooks/data/fulldata_03_11_03_P_숙박업.csv'
df = pd.read_csv(fpath, encoding='cp949',low_memory=False)
pd.options.display.max_columns = None
df['year'] =  df['최종수정시점'].apply(lambda x: x.split('-')[0]).astype(np.int64)
new_df = df.groupby(['상세영업상태명','year'])['번호'].count()
new_df['영업'].index, new_df['영업'].values
```

```python
plt.figure(figsize=(15,5))
plt.plot(new_df['영업'].index,new_df['영업'].values, label='영업')
plt.plot(new_df['폐업'].index,new_df['폐업'].values, label='폐업')
plt.title("숙박업소 - 영업")
plt.xlabel("년도")
plt.ylabel("개수")
plt.legend(loc = 'upper left')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.plot(new_df['영업'].index, np.log10( new_df['영업'].values) )
plt.title("숙박업소 - 영업")
plt.show()
```
