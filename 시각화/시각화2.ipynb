```python
# 산점도(scatter plot) : 데이터 분포를 2차원 평면에 표현
# c : 포인트 색상
# maker는 포인트 모양
# size는 포인트 크기
# alpha는 불투명
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_1 = np.random.rand(512,2)
data_2 = np.random.rand(512,2)
```

```python
data_1
```

```python
plt.scatter(data_1[:, 0], data_1[:, 1]  , c='b', marker='x')
plt.scatter(data_2[:, 0], data_2[:, 1]  , c='r', marker='o')
plt.show()
```

```python
plt.scatter(data_1[:, 0], data_1[:, 1]  , c='b', marker='x')
plt.scatter(data_2[:, 0], data_2[:, 1]  , c='r', marker='o',alpha = 0.5)
plt.show()
```

```python
# iris 붓꽃
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
# 품종이 'setosa' 데이터만 추출
setosa = iris.loc[ iris['species'] == 'setosa' ]
setosa.head()
```

```python
plt.scatter(setosa['sepal_length'], setosa['sepal_width'])
```

```python
# species(품종) 의 종류는 ?
iris['species'].value_counts()
```

```python
# 품종별로 데이터 추출
setosa = iris.loc[ iris['species'] == 'setosa' ]
versicolor = iris.loc[ iris['species'] == 'versicolor' ]
virginica  = iris.loc[ iris['species'] == 'virginica' ]
```

```python
plt.scatter(setosa['petal_length'], setosa['petal_width'], label = 'setosa')
plt.scatter(versicolor['petal_length'], versicolor['petal_width'],label = 'versicolor')
plt.scatter(virginica['petal_length'], virginica['petal_width'], label = 'virginica')
plt.legend()
plt.show()
```

```python
# plot --> 점으로 이루어진 선.... 시간의 흐름. 즉 시간정보(과거와 현재 또는 앞 뒤)를 가지고 있는 데이터 시계열
# 시계열 : 주식, 기온변화에따른.. , 문장
```
