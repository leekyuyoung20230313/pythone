```python
from bs4 import BeautifulSoup
import urllib.request
```

```python
pdResult = []
```

```python
import pandas as pd
from tqdm import tqdm
for pagenum in tqdm(range(1,100)):
  try:    
    url = f'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={pagenum}'
    print(url)
    html = urllib.request.urlopen(url)
    soup = BeautifulSoup(html,'html.parser')
    tag_all_tr = soup.select("div.tableType01 > table > tbody > tr")
    result = []
    for tag in tag_all_tr:
      tag_td_all = tag.find_all('td')    
      templist = []
      if len(tag_td_all) < 2:
        raise Exception('등록된 매장이 없습니다.')
      for i in tag_td_all:
        templist.append(i.string)    
      result.append(templist)
    pdResult.append(pd.DataFrame(result))
  except Exception as e:
    print(e)
    break
```

```python
df = pd.concat(pdResult)
df = df.reset_index(drop=True)
df.head()
```

```python
df.columns = ['지역'	,'매장명',	'현황',	'주소',	'매장 서비스'	,'전화번호']
df = df.set_index('매장명', drop=True)
df.head()
```

```python
df = df.drop(['매장 서비스'], axis=1)
df.head()
```

```python
df.to_csv('./coffe1.csv',encoding='utf-8')
```

```python
# 매장서비스 추가
pdResult=[]
import pandas as pd
for pagenum in range(1,2):
  try:    
    url = f'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={pagenum}'    
    html = urllib.request.urlopen(url)
    soup = BeautifulSoup(html,'html.parser')
    tag_all_tr = soup.select("div.tableType01 > table > tbody > tr")
    result = []
    for tag in tag_all_tr:
      tag_td_all = tag.find_all('td')    
      templist = []
      if len(tag_td_all) < 2:
        raise Exception('등록된 매장이 없습니다.')
      for idx, i in enumerate(tag_td_all):
        if idx == 4:
          imglist = i.find_all('img')
          templist.append(','.join([img.attrs['alt'] for img in imglist]))
        else:
          templist.append(i.string)    
      result.append(templist)
    pdResult.append(pd.DataFrame(result))
  except Exception as e:
    print(e)
    break
```
