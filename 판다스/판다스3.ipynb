```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# 데이터 로드
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# 성별 평균나이
# 분할 - 적용 - 합
titanic.groupby(['sex'])['age'].mean()
```

```python
# 객실 등급별 승객수  count()  갯수를  세는 함수
titanic.groupby(['pclass'])['sex'].count()
```

```python
# 객실별, 성별, 승객수
titanic.groupby(['pclass','sex'])['sex'].count()
```
