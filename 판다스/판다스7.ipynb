```python
# 타이타닉데이터로 연습
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic
# survived	pclass	sex	age
# 1 survived	pclass	sex	age  이 컬럼만으로 구성된 새로운 데이터 프레임
# 2. info()를 이용해서 결측치(na) 및 데이터 타입을 조회
# 3 na가 있는지 확인하고 na가 있는 모든 항목을 출력  데이터프레임객체.isna() 이용
# 4 na가 있는 항목들을 적절하게 값을 채운다 - transform함수를이용.... chat gpt로 검색
```

```python
# 1 
new_titanic = titanic[['survived',	'pclass',	'sex',	'age']]
new_titanic.head()
```

```python
# 2.
new_titanic.info()
```

```python
# 3.
idx = new_titanic.isna().any(axis=1)
idx
```

```python
new_titanic[idx]
```

```python
# 4 
new_titanic.groupby(['pclass',	'sex'])['age'].transform(lambda x : x.fillna(x.mean()))
```

```python
# new_titanic.loc[:,'age'] =  new_titanic.groupby(['pclass',	'sex'])['age'].transform(lambda x : x.fillna(x.mean()))
```

```python
new_titanic.loc[new_titanic[idx].index,'age']= new_titanic.groupby(['pclass',	'sex'])['age'].transform(lambda x : x.fillna(x.mean()))
```
