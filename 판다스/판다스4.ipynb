```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# 데이터 로드
titanic = sns.load_dataset('titanic')
titanic.head()  # 데이터를 간략하게 볼때
```

```python
# 데이터 타입...
titanic.info()
```

```python
# 데이터를 재 구성
titanic.head()
```

```python
new_titanic = titanic.loc[:,:'age']
```

```python
new_titanic.info()
```

```python
# dropna() na 항목을 전부 제거
```

```python
new_titanic.head(10)
```

```python
new_titanic.groupby(['sex']).transform(lambda x: x.max())
```

```python
new_titanic.groupby(['sex']).transform(lambda x: 100)
```

```python
new_titanic.groupby(['pclass',	'sex'])['age'].transform(lambda x: x.mean())
```

```python
new_titanic.head()
```

```python
new_titanic['age'] = new_titanic.groupby(['pclass','sex'])['age'].transform(lambda x: x.fillna(x.mean()))
```

```python
new_titanic.info()
```

```python
new_titanic['age'] = new_titanic['age'].astype(np.int16)
```

```python
new_titanic.info()
```

```python
new_titanic['newdata'] = '23,568'
```
