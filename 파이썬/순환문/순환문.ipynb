```python
# 순환문..... 순환(조건을 만족하면)
# 일정패턴의 반복되는 코드
# 사전만들기
# 5번 돌면서 사전 작성
mydic = {}
i = 0
while i < 5:
  word = input('찾고자 하는 단어를 입력하세요 : ')
  if word not in mydic.keys():
    print("찾고자 하는 단어어 없습니다. 값을 입력해서 사전에 추가해 주세요")
    mydic[word] = input("단어의 뜻을 입력하세요 : ")
  else:
    print(mydic[word])
    
  i += 1
```

```python
# 1과 10사이의 랜덤하게 데이터를 10개 리스트에 저장
import random
mylist = []
```

```python
i = 0
while i < 10:
  mylist.append(random.randrange(1,100))
  i += 1
```

```python
# 주어진 리스트에서 가장 큰 값을 찾아서 출력
import random
mylist=[]
i = 0
while i < 10:
  mylist.append(random.randrange(1,100))
  i += 1
mylist  
```

```python
# 쵀대값 또는 최소값을 찾아서 출력
# 리스트의 첫번째값을 최대값을 저장하는 변수의 값으로 초기화
# 리스트의 데이터를 2번째부터 전부순환하면서
# 각각의 값과 최대값 변수를 비교해서 리스트의 값이 크면 최대값 변수를 없데이트
mymax = mylist[0]
i = 1
while i < len(mylist):
  if mylist[i] > mymax:
    mymax = mylist[i]
  i += 1
print(f"최대값은 {mymax}")
```

```python
# 숫자를 입력받아서 정수로 변환.....
# 아래코드를 while문으로 5번 순환하도록 개선하고... 만약에 정상적으로 입력하면
# 남은 순환횟수에 관계없이 순환문을 종료하세요

number = input("숫자를 입력하세요 : ")
if number.isdigit():
  number = int(number)
  number + 10
else:
  number = input("숫자를 입력하세요 : ")
  if number.isdigit():
    number = int(number)
    number + 10
  else:
    number = input("숫자를 입력하세요 : ")
```
